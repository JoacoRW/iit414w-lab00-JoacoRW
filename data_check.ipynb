```python
# -- Reproducibility Header ------------------------------------------------
import sys
import random
import numpy as np
import warnings

RANDOM_SEED = 414
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
warnings.filterwarnings('ignore', category=FutureWarning)

print(f'Python  : {sys.version.split()[0]}')
print(f'NumPy   : {np.__version__}')
print(f'Seed    : {RANDOM_SEED}')
```

```python
# -- FastF1 Session Load ----------------------------------------------------
import os
import fastf1

cache_path = os.path.abspath(os.path.join(os.getcwd(), 'data', 'fastf1_cache'))
os.makedirs(cache_path, exist_ok=True)
fastf1.Cache.enable_cache(cache_path)

session = fastf1.get_session(2024, 'Bahrain', 'R')
session.load(laps=True, telemetry=False, weather=False, messages=False)

print(f"Session: {session.name}")
print(f"Event:   {session.event['EventName']}")
print(f"Results shape: {session.results.shape}")
print(f"Results columns: {list(session.results.columns)}")
print('First 5 lap rows:')
session.laps.head()
```

```python
# -- Jolpica API Query ------------------------------------------------------
import json
from urllib.request import urlopen

url = 'https://api.jolpi.ca/ergast/f1/2024/driverstandings/'

with urlopen(url) as response:
    status_code = response.status
    payload = json.load(response)

driver_standings = payload['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']

print(f'HTTP status code: {status_code}')
print(f'Number of records returned: {len(driver_standings)}')
print('First 3 driver entries (name, nationality, points):')

for entry in driver_standings[:3]:
    driver = entry['Driver']
    full_name = f"{driver['givenName']} {driver['familyName']}"
    nationality = driver['nationality']
    points = entry['points']    
    print(f'- {full_name} | {nationality} | {points}')
```

## Data Shape Summary

- **FastF1 session chosen and why:** I chose 2024 Bahrain Race because it is a complete race session with stable, well-known data coverage.
- **Shapes of both datasets:**
  - FastF1 results shape: `session.results.shape` (fill after run, expected around 20 drivers x many columns).
  - Jolpica driver standings records: `len(driver_standings)` (fill after run).
- **One observation:** FastF1 returns rich, race-engineering style tables with many timing columns, while Jolpica provides lighter JSON focused on standings and metadata.
